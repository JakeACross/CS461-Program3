```python
import re
import pandas as pd
import numpy as np
```

```python
def normalize(df, name):
    max_value = df[name].max()
    min_value = df[name].min()
    result = (df[name] - min_value) / (max_value - min_value)
    return result
```

```python
# Load dataset
data = pd.read_csv('../congressional_tweet_training_data.csv')

data.head(10)
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop = stopwords.words('english')

hashtags = data['hashtags'].values
hashtag = []
for e in hashtags:
    if len(e.split()) > 1:
        hashtag.extend(e.split())
    else:
        hashtag.append(e)
```

```python
data['full_text'] = data['full_text'].apply((lambda x: re.sub('#[a-zA-z0-9]*','',x)))  # remove hashtag
data['full_text'] = data['full_text'].apply((lambda x: re.sub('[^a-zA-z0-9\s]','',x)))  # Exclude punctuations
data['full_text'] = data['full_text'].apply(lambda x: x[1:])  # remove b
data['full_text'] = data['full_text'].apply(lambda x: x.replace('rt',''))  # reomove rt
data['full_text'] = data['full_text'].apply(lambda x: x.lower())  # get lowercase
data['full_text'] = data['full_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop]))  # remove stop words

data['full_text'][2]
```

```python
texts = data['full_text'].values
length = [len(text) for text in texts]
```

https://stackoverflow.com/questions/48784908/numpy-unique-sort-based-on-counts

```python
ht, count = np.unique(hashtag, return_counts=True)
sorted_ind = np.argsort(-count)

print(ht[sorted_ind])
print(count[sorted_ind])

top_hashtag = ht[sorted_ind][:500]
top_hashtag = np.append(top_hashtag, 'Other')

new_columns = {}
for k in top_hashtag:
    new_columns[k] = []
print(new_columns)
```

```python
count = 0
for tags in hashtags:
    count = 0
    max_len = 0
    if len(tags.split()) > 1:
        for tag in tags.split():
            if tag in new_columns:
                new_columns[tag].append(1)
                max_len = len(new_columns[tag])
            elif count == 0:
                new_columns['Other'].append(1)
                count += 1
                max_len = len(new_columns['Other'])
            
    else:
        if tags in new_columns:
            new_columns[tags].append(1)
            max_len = len(new_columns[tags])
        else:
            new_columns['Other'].append(1)  
            max_len = len(new_columns['Other'])
    
    for key in new_columns:
        if len(new_columns[key]) < max_len:
            new_columns[key].append(0)  
            
sum(new_columns['COVID19'])
```

```python
data = data.drop(columns = ['full_text', 'hashtags', 'year'])

data.insert(1, 'length', length, True)

insert_ind = 2
for k in new_columns:
    data.insert(insert_ind, k, new_columns[k], True)
    insert_ind += 1
```

```python
data['favorite_count'] = normalize(data, 'favorite_count')
data['length'] = normalize(data, 'length')
data['retweet_count'] = normalize(data, 'retweet_count')
```
